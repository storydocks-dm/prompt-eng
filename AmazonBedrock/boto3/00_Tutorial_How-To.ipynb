### Anthropic SDK 与 Messages API
在本教程中，我们将使用 [Anthropic python SDK](https://docs.anthropic.com/claude/reference/client-sdks) 和 [Messages API](https://docs.anthropic.com/claude/reference/messages_post)。

下面是本教程中运行提示的示例。首先，我们创建 `get_completion`，这是一个辅助函数，它将提示发送给 Claude 并返回 Claude 生成的响应。现在运行该单元格。

首先，我们设置并存储模型名称和区域。

```python
import boto3
session = boto3.Session() # create a boto3 session to dynamically get and set the region name
AWS_REGION = session.region_name
print("AWS Region:", AWS_REGION)
MODEL_NAME = "anthropic.claude-3-haiku-20240307-v1:0"

%store MODEL_NAME
%store AWS_REGION
```

然后，我们创建 `get_completion`，这是一个辅助函数，它将提示发送给 Claude 并返回 Claude 生成的响应。现在运行该单元格。

```python
import boto3
import json

bedrock = boto3.client('bedrock-runtime',region_name=AWS_REGION)

def get_completion(prompt):
    body = json.dumps(
        {
            "anthropic_version": '',
            "max_tokens": 2000,
            "messages": [{"role": "user", "content": prompt}],
            "temperature": 0.0,
            "top_p": 1,
            "system": ''
        }
    )
    response = bedrock.invoke_model(body=body, modelId=MODEL_NAME)
    response_body = json.loads(response.get('body').read())

    return response_body.get('content')[0].get('text')
```

现在我们将为 Claude 编写一个示例提示，并通过运行我们的 `get_completion` 辅助函数打印 Claude 的输出。运行下面的单元格将在其下方打印 Claude 的响应。

请随意调整提示字符串以从 Claude 获得不同的响应。

```python
# Prompt
prompt = "Hello, Claude!"

# Get Claude's response
print(get_completion(prompt))
```

前面定义的 `MODEL_NAME` 和 `AWS_REGION` 变量将在整个教程中使用。只需确保从上到下运行每个教程页面的单元格即可。
