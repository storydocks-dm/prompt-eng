# 附录 10.3：搜索与检索

您知道吗，您可以使用 Claude **为您搜索维基百科**？Claude 可以查找和检索文章，之后您还可以使用 Claude 来总结和综合它们，从它找到的内容中编写新颖的内容等等。不仅仅是维基百科！您还可以搜索您自己的文档，无论是作为纯文本存储还是嵌入在向量数据存储中。

查看我们的 [RAG cookbook 示例](https://github.com/anthropics/anthropic-cookbook/blob/main/third_party/Wikipedia/wikipedia-search-cookbook.ipynb)以了解如何使用从向量数据库、维基百科、互联网等检索的数据来补充 Claude 的知识并提高 Claude 响应的准确性和相关性。在那里，您还可以了解如何使用某些[嵌入](https://docs.anthropic.com/claude/docs/embeddings)和向量数据库工具。

如果您有兴趣了解使用 Claude 的高级 RAG 架构，请查看我们关于 [RAG 架构的 Claude 3 技术演示幻灯片](https://docs.google.com/presentation/d/1zxkSI7lLUBrZycA-_znwqu8DDyVhHLkQGScvzaZrUns/edit#slide=id.g2c736259dac_63_782)。
