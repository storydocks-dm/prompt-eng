### Anthropic SDK 和 Messages API
我们将在整个教程中使用 [Anthropic python SDK](https://docs.anthropic.com/claude/reference/client-sdks) 和 [Messages API](https://docs.anthropic.com/claude/reference/messages_post)。

下面是本教程中运行提示的示例。首先，我们创建 `get_completion`，这是一个辅助函数，它将提示发送给 Claude 并返回 Claude 生成的响应。现在运行该单元格。

```python
import anthropic

client = anthropic.Anthropic(api_key=API_KEY)

def get_completion(prompt: str):
    message = client.messages.create(
        model=MODEL_NAME,
        max_tokens=2000,
        temperature=0.0,
        messages=[
          {"role": "user", "content": prompt}
        ]
    )
    return message.content[0].text
```

现在我们将为 Claude 编写一个示例提示，并通过运行 `get_completion` 辅助函数打印 Claude 的输出。运行下面的单元格将在其下方打印来自 Claude 的响应。

您可以随意修改提示字符串，以从 Claude 获得不同的响应。

```python
# Prompt
prompt = "Hello, Claude!"

# Get Claude's response
print(get_completion(prompt))
```

之前定义的 `API_KEY` 和 `MODEL_NAME` 变量将在整个教程中使用。只需确保每个教程页面的单元格从上到下运行即可。
